```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split #We need this to split the data
import seaborn as sns
```

```python
df = pd.read_csv("CompleteDataset.csv")
df.head()
```

```python
#Check the data available
df.info()
```

### Remove the unneeded preferred positions

```python
# Extract the first Preferred Position
split = df["Preferred Positions"].str.split(" ", n = 1, expand = True) 
  
# making separate column for Preferred Position
df["Preferred Position"]= split[0] 

```

### Subset needed data

```python
# subseting the Dataframe
df_subset = df[["Preferred Position", "Acceleration","Aggression", "Agility", "Balance", "Ball control", "Composure", "Crossing", "Curve", "Dribbling", "Finishing", "Free kick accuracy", "GK diving", "GK handling", "GK kicking", "GK positioning", "GK reflexes", "Heading accuracy", "Interceptions", "Jumping", "Long passing", "Long shots", "Marking", "Penalties", "Positioning", "Reactions", "Short passing", "Shot power", "Sliding tackle", "Sprint speed", "Stamina", "Standing tackle", "Strength", "Vision", "Volleys"]]
df_subset.head()
```

##### Check data and run KNN

```python
df_subset["Preferred Position"].value_counts()
```

```python
y = df_subset["Preferred Position"]
X = df_subset[["Acceleration","Aggression", "Agility", "Balance", "Ball control", "Composure", "Crossing", "Curve", "Dribbling", "Finishing", "Free kick accuracy", "GK diving", "GK handling", "GK kicking", "GK positioning", "GK reflexes", "Heading accuracy", "Interceptions", "Jumping", "Long passing", "Long shots", "Marking", "Penalties", "Positioning", "Reactions", "Short passing", "Shot power", "Sliding tackle", "Sprint speed", "Stamina", "Standing tackle", "Strength", "Vision", "Volleys"]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier #the object class we need
knn = KNeighborsClassifier(n_neighbors=5) #create a KNN-classifier with 5 neighbors (default)
knn = knn.fit(X_train, y_train) #this fits the k-nearest neigbor model with the train data
knn.score(X_test, y_test) #calculate the fit on the test data
```

##### Our overall KNN score is .4970 so we are correctly predicting position in ~50% of cases. Now lets look deeper at the precision and recal per position

```python
from sklearn.metrics import classification_report
y_pred = knn.predict(X_test) 
print(classification_report(y_test, y_pred))
```

#### We can clearly see that the algorithim is better at picking some positions more than others. 

For example Goalkeepers were flawlessly predicted by the model. This is unsurprising when you consider that Goalkeepers has significantly higher stats than outfield players in relation to GK values. 

Additionally, Strikers were predicted at a relativly high rate. This is again likely due to the high reliance the position has on doing one job, scoring goals. This is very different for example in comparison to a midfielder, who could be a defensive player, attacking second striker, or a creative player. Couple this with the fact that these players are split across CF, CAM, CM and CDM. 

#### Maybe its better to use a random forrest algorithim instead?

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=1, n_estimators=100) #RF is a random algorithm, so to get the same results we need to use random_state
rf = rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test) #the predicted values
print(classification_report(y_test, y_pred))
```

#### The random forrest results are marginally better than the KNN output. An much higher average precision and marginally higher recall.

However the change is minimal, while it's better in some ways (the LWB prediction got all LWB players!) its worse in others (it only got 3% of them...).
